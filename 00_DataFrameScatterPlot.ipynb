```python
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd
from pandas import Series, DataFrame

```

```python
# Scatter plot form a random created data

N = 60 #random number range

grp1 = (0.6 + 0.6 * np.random.rand(N), np.random.rand(N))
grp2 = (0.4+0.3 * np.random.rand(N), 0.5*np.random.rand(N))
grp3 = (0.3*np.random.rand(N),0.3*np.random.rand(N))

data = (grp1, grp2, grp3)
colors = ("red", "green", "yellow")
groups = ("coffee", "juice", "water")

# Create plot
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, facecolor="1.0")

for data, color, group in zip(data, colors, groups):
    x, y = data 
    ax.scatter(x, y, alpha=0.8, c=color, edgecolors='none', s=30, label=group)

plt.title('Matplot scatter plot')
plt.legend(loc=0)
plt.show()
```
